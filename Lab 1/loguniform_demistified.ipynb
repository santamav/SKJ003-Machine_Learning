```python
import numpy as np
import pandas as pd
import random
import seaborn as sns
import hyperopt
from hyperopt import hp
```

```python
values = []
space = hp.loguniform('something', 0, 1)
for x in range(1000):
    values.append(hyperopt.pyll.stochastic.sample(space))
df = pd.DataFrame(values, columns=["var1"])
df.hist(bins=100)
```

```python
values = []
space = hp.loguniform('something', np.log(0.001), np.log(0.01))
for x in range(1000):
    values.append(hyperopt.pyll.stochastic.sample(space))
df = pd.DataFrame(values, columns=["var1"])
df.hist(bins=100)
```

```python
import numpy as np
import math
import matplotlib.pyplot as plt

ini = np.log(1e-3)
end = np.log(1e-2)
n_samples = 100
step = (end - ini) / n_samples
exp_range = np.arange(ini, end, step)

lr = np.exp(exp_range)

plt.plot(exp_range, lr)
plt.xlim(ini, end)
plt.ylim(1e-3, 1e-2)
plt.axvline(x=ini+(end-ini)/2, color = 'r', linestyle = '-')
plt.show()
```
